```python
import pandas as pd

# CSVファイルからDataFrameを読み込む
dataset = pd.read_csv('/content/airpollutiondata.csv', index_col=0)

# インデックスをリセット
dataset.reset_index(inplace=True)

# Timestamp列を削除
dataset.drop('Timestamp', axis=1, inplace=True)

dataset = dataset.fillna(0)


# 100以上の値を0に置き換える
dataset[dataset >= 100] = 0

# 結果を表示
dataset.head()
```

```python
maxValueInEachColumn = dataset.max() #Gathering the maximum PM2.5 value recorded by each sensor
maxValueInEachColumn      #Printing the maxValue of each sensor
```

```python
minValueInEachColumn = dataset.min() #Reading the minimum PM2.5 value recorded by each sensor
minValueInEachColumn.plot()  #ploting them

#dataset.min().plot()      #memory efficient approach
```

```python
import pandas as pd
import plotly.express as px
import re
import plotly.io as pio

# Colabレンダラーを設定
pio.renderers.default = 'colab'

# ファイルパス
file_path = 'frequentPatterns.txt'

# ファイルの読み取り
data = []
with open(file_path, 'r') as file:
    for line in file:
        # 行から座標と頻度を抽出
        matches = re.findall(r'POINT\(([^)]+)\):(\d+)', line)
        if matches:
            for match in matches:
                coords, freq = match
                lon, lat = map(float, coords.split())  # 座標の順序を修正
                freq = int(freq)
                data.append((lat, lon, freq))

# データフレームの作成
df = pd.DataFrame(data, columns=['Latitude', 'Longitude', 'Frequency'])

# 最長のパターンを特定
most_frequent = df.groupby(['Latitude', 'Longitude']).sum().reset_index()
most_frequent = most_frequent.sort_values(by='Frequency', ascending=False).iloc[0]

# 最長パターンのみをフィルタリング
most_frequent_df = pd.DataFrame([most_frequent])

# 最長パターンを表示
print(f"Most frequent pattern: Latitude = {most_frequent['Latitude']}, Longitude = {most_frequent['Longitude']}, Frequency = {most_frequent['Frequency']}")

# 地図上にプロット
fig = px.scatter_mapbox(most_frequent_df,
                       lat='Latitude',
                       lon='Longitude',
                       size='Frequency',
                       hover_name='Frequency',
                       mapbox_style='open-street-map',
                       title='Most Frequent Pattern on OpenStreetMap')

# プロットの設定
fig.update_layout(margin={"r": 0, "t": 0, "l": 0, "b": 0})
fig.update_layout(title_text="Most Frequent Pattern Location")

# グラフの表示
fig.show()
fig.write_html('Frequent Pattern')
```
